```python
import pandas as pd
```

```python
df = pd.read_csv(r"first_url_data.csv")
df.head()
```

```python
geo_coordinate = pd.read_csv(r"Geospatial_Coordinates.csv")
geo_coordinate.columns = ['PostalCode', 'Latitude', 'Longitude']
geo_coordinate.head()
```

```python
merge_data = pd.merge(df, geo_coordinate,on = 'PostalCode')
merge_data.head(20)
```

```python
merge_data.to_csv("second_url_data.csv", index = False)
```
